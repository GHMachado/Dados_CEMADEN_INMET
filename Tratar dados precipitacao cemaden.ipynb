```python
import pandas as pd

# Abra o arquivo CSV com o separador correto (assumindo que seja vírgula)
df = pd.read_csv('/content/drive/MyDrive/meteorologia_dados/dados_prec/data.csv', sep=';') # Substitua 'nome_do_arquivo.csv' pelo nome real do seu arquivo

# Exclua as colunas 2, 3, 5 e 6 (colunas 1, 2, 4 e 5 no índice do Python)
df = df.drop(df.columns[[1, 2, 4, 5]], axis=1)

# Renomeie as colunas para facilitar o entendimento
df = df.rename(columns={df.columns[0]: 'municipio', df.columns[1]: 'nomeEstacao', df.columns[2]: 'datahora', df.columns[3]: 'valorMedida'})

# Substituir virgulas por pontos e converte para Float
df['valorMedida'] = df['valorMedida'].str.replace(',', '.').astype(float)

# Converta a coluna 'DataHora' para datetime
df['datahora'] = pd.to_datetime(df['datahora'])

# Agrupe por estação, município e data (dia) e calcule o acumulado diário
df_diario = df.groupby(['nomeEstacao', 'municipio', df['datahora'].dt.date])['valorMedida'].sum().reset_index()

# Crie a tabela pivotada com dias em colunas e estações em linhas, incluindo o município
tabela_pivotada = df_diario.pivot(index=['nomeEstacao', 'municipio'], columns='datahora', values='valorMedida')

# Concatenar nomeEstacao e municipio para uma melhor visualização
tabela_pivotada.index = tabela_pivotada.index.map(lambda x: f"{x[0]} - {x[1]}")

# Salva como uma tabela de excel
tabela_pivotada.to_csv('/content/drive/MyDrive/meteorologia_dados/dados_prec/tabela_precipitacao_FEV.csv', sep=';', decimal='.')

# Exiba a tabela pivotada
# Para ver a saída, execute o código.
print(tabela_pivotada)
```
